```python
import os
filenames=os.listdir("/home/wsuser/work/Dataset Plant Disease/Veg-dataset/Veg-dataset/test_set")
```

```python
x_train=train_datagen.flow_from_directory('/home/wsuser/work/Dataset Plant Disease/Veg-dataset/Veg-dataset/test_set',target_size=(128,128),batch_size=2,class_mode='categorical')
x_test=test_datagen.flow_from_directory(r'/home/wsuser/work/Dataset Plant Disease/Veg-dataset/Veg-dataset/train_set/',target_size=(128,128),batch_size=2,class_mode='categorical')
```

```python
x_train.class_indices
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
```

```python
model=Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape=(128,128,3),activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units=300,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(units=150,kernel_initializer='random_uniform',activation='relu'))
```

```python
model.add(Dense(units=75,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(units=9,kernel_initializer='uniform',activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(x_train,steps_per_epoch=89,epochs=20,validation_data=x_test,validation_steps=27)
```
